```python
import random
import numpy as np
import torch
import torch.backends.cudnn as cudnn

import matplotlib.pyplot as plt

# Set the seed for reproducibility
seed = 42
cudnn.benchmark = False
cudnn.deterministic = True

random.seed(seed)
torch.manual_seed(seed)
if torch.cuda.is_available():
    torch.cuda.manual_seed(seed)
np.random.seed(seed)
```

```python
from factory.trainer import Trainer
from sklearn.metrics import (
    confusion_matrix,
    accuracy_score,
    precision_score,
    recall_score,
    f1_score
)
```

# Training

```python
trainer = Trainer()
trainer.run()
```

## Training results

```python
# Compute the accuracy
accuracy = accuracy_score(trainer.y_test, trainer.y_pred)
f1 = f1_score(trainer.y_test, trainer.y_pred, average ="weighted")
precision = precision_score(trainer.y_test, trainer.y_pred, average ="weighted")
recall = recall_score(trainer.y_test, trainer.y_pred, average ="weighted")

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,8))
sns.set(font_scale = 1.5)

conf_matrix = confusion_matrix(trainer.y_test, trainer.y_pred)


ax = sns.heatmap(
    conf_matrix, # confusion matrix 2D array 
    annot=True, # show numbers in the cells
    fmt='d', # show numbers as integers
    cbar=False, # don't show the color bar
    cmap='Blues', # customize color map
    vmax=175 # to get better color contrast
)

ax.set_xlabel("Predicted", labelpad=20)
ax.set_ylabel("Actual", labelpad=20)
plt.show()
```

```python
interpretation_summary = trainer.interpretation_summary
print(interpretation_summary.keys())

y_true = interpretation_summary['y_true']
y_gaussian = interpretation_summary['gaussian']
```

```python
print('Accuracy: ', accuracy_score(y_true, y_gaussian))
print('Precision: ', precision_score(y_true, y_gaussian, average='weighted'))
print('Recall: ', recall_score(y_true, y_gaussian, average='weighted'))
print('F1: ', f1_score(y_true, y_gaussian, average='weighted'))
```
